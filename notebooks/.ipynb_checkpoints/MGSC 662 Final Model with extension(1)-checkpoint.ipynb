# Model

```python
import gurobipy as gb
from gurobipy import *
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import math
```

```python
# Phone locations (current + candidates)
df_candidate = pd.read_csv("exisiting_and_candidats_callbox.csv")
candidate_x = df_candidate['x'].to_list()
candidate_y = df_candidate['y'].to_list()
J = []

for i in range(len(df_candidate)):
  J.append((candidate_x[i],candidate_y[i]))

J[:5]
```

```python
# Demand points
df_demand = pd.read_csv("demand_points_combined.csv")
demand_x = df_demand['x'].to_list()
demand_y = df_demand['y'].to_list()
I = []

for i in range(len(df_demand)):
  I.append((demand_x[i],demand_y[i]))
```

```python
len(I)
```

```python
# Current state of phones
df_current = pd.read_csv("current_state.csv")
print(f"Loaded current state for {len(df_current)} existing locations")

# Dictinaries
phone_names = {
    0: "Type 1 Panoramic",
    1: "Type 1 Fixed",
    2: "Type 2",
    3: "Type 3",
    4: "Old Phone"
}

condition_names = {0: "New", 1: "Good", 2: "Rusted", 3: "Broken"}
```

```python
demand_num = len(I)
location_num = len(J)
k_num = 5
period_num = 2

# Euclidian distance matrix calculation
I_arr = np.array(I)
J_arr = np.array(J)
dist_matrix = np.sqrt(((I_arr[:, None, :] - J_arr[None, :, :])**2).sum(axis=2))

R = [150, 100, 60, 45, 20]  # Coverage radius for each phone type
C = [19_500, 10_500, 3500, 2000, 0]  # Cost (k=4 has 0 cost as it's keeping old)
lmbda = 1000  # Penalty multiplier for not replacing
alpha = 0.75  # % of demand points to cover
W = df_demand['weight'].to_list()  # Weights for each demand location

# Condition of existing phones (0=new, 1=good, 2=rusted, 3=broken)
H_old = df_current['State H'].to_list()
H = H_old + [0] * (location_num - len(H_old))  # No phones at new locations

# Extract initial phone types and conditions
initial_phone_types = df_current['Phone K'].to_list()

# Create model
model = gb.Model('Emergency Phone Box Optimization')
```

### Decision Variables

```python
# Decision variables
X = model.addVars(period_num, location_num, k_num, vtype=GRB.BINARY)
A = model.addVars(demand_num, location_num, k_num, vtype=GRB.BINARY)
Y = model.addVars(demand_num, vtype=GRB.BINARY)
```

### Objective Function

```python
# Cost of installing NEW phones (k=0,1,2,3) at ANY location in period 1
# For existing locations (j < 23): only charge if replacing (k != existing type)
installation_cost_existing = gb.quicksum(
    X[1, j, k] * C[k]
    for j in range(23)
    for k in range(4)
    if k != initial_phone_types[j]
)

# For new locations (j >= 23): always charge equipment cost
installation_cost_new = gb.quicksum(
    X[1,j,k] * C[k]
    for j in range(23, location_num)
    for k in range(4)
)

# Penalty for KEEPING old phones at EXISTING locations
# Where condition: 0=new, 1=good, 2=rusted, 3=broken
penalty_keep_old = gb.quicksum(
    X[1,j,4] * H[j] * lmbda
    for j in range(23)
)

# Set objective
model.setObjective(installation_cost_existing + installation_cost_new + penalty_keep_old, GRB.MINIMIZE)

model.update()
```

### Constraints

```python
# 1. INITIAL STATE CONSTRAINTS (Period t = 0)
# Existing locations (0-22) have old phones initially
for j in range(len(initial_phone_types)):
    k_existing = initial_phone_types[j]
    
    # This location has phone type k_existing
    model.addConstr(X[0, j, k_existing] == 1, name=f"init_phone_{j}_{k_existing}")
    
    # All other phone types are 0 at this location
    for k in range(k_num):
        if k != k_existing:
            model.addConstr(X[0, j, k] == 0, name=f"no_phone_{j}_{k}")

# New locations (23-122) have nothing initially
for j in range(23, location_num):
    for k in range(k_num):
        model.addConstr(X[0,j,k] == 0, name=f"empty_initial_{j}_{k}")

# 2. PERIOD 1 CONSTRAINTS
# At most one phone type per location in period 1
model.addConstrs(
    gb.quicksum(X[1,j,k] for k in range(k_num)) <= 1 
    for j in range(location_num)
)

# New locations (23-122) cannot have old phones in period 1
for j in range(23, location_num):
    model.addConstr(X[1,j,4] == 0, name=f"no_old_new_loc_{j}")

# 3. REPLACEMENT LOGIC
# If we keep the old phone in period 1, it must have existed in period 0. This is automatically satisfied by initial state constraints

# 4. COVERAGE CONSTRAINTS
# A demand point can only be assigned to a phone that exists in period 1
model.addConstrs(
    A[i,j,k] <= X[1,j,k]
    for i in range(demand_num)
    for j in range(location_num)
    for k in range(k_num)
)

# Each demand point is covered if assigned to at least one phone
model.addConstrs(
    gb.quicksum(A[i,j,k] for j in range(location_num) for k in range(k_num)) >= Y[i] 
    for i in range(demand_num)
)

# Coverage radius constraints
for i in range(demand_num):
    for j in range(location_num):
        for k in range(k_num):
            if dist_matrix[i, j] > R[k]:
                model.addConstr(A[i, j, k] == 0, name=f"radius_{i}_{j}_{k}")

# Minimum coverage requirement
model.addConstr(
    gb.quicksum(W[i]*Y[i] for i in range(demand_num)) >= 
    alpha * gb.quicksum(W[i] for i in range(demand_num))
)

# 5. NO REMOVAL - existing locations must always have a phone
for j in range(23):
    model.addConstr(
        gb.quicksum(X[1,j,k] for k in range(k_num)) == 1, 
    )

# 6. NO DOWNGRADE - cannot replace with worse phone type
# Phone quality hierarchy: 0 (best) > 1 > 2 > 3 > 4 (worst)
for j in range(23):
    k_existing = initial_phone_types[j]
    
    # Cannot install phones with higher k value (worse quality)
    # Exception: Can always upgrade from old phone (k=4) to any new type
    if k_existing < 4:  # If it's not an old phone
        for k_worse in range(k_existing + 1, k_num):
            model.addConstr(
                X[1,j,k_worse] == 0, 
                name=f"no_downgrade_{j}_to_{k_worse}"
            )
```

```python
# Optimize
model.optimize()

# Print results
if model.status == GRB.OPTIMAL:
    print(f"Optimal Solution: ${model.objVal:,.2f}\n")
    
    # Detailed decisions for existing locations
    print("DECISIONS FOR EXISTING LOCATIONS:")
   
    replaced = []
    kept = []
    
    for j in range(len(initial_phone_types)):
        k_existing = initial_phone_types[j]
        condition = H_old[j]
        
        # Check what happened at this location
        kept_existing = X[1, j, k_existing].x > 0.5
        
        if kept_existing:
            kept.append(j)
            print(f"Location {j:2d}: KEEP {phone_names[k_existing]:<20} (Condition: {condition_names[condition]})")
        else:
            # Find what it was replaced with
            for k_new in range(k_num):
                if k_new != k_existing and X[1, j, k_new].x > 0.5:
                    replaced.append((j, k_existing, k_new))
                    print(f"Location {j:2d}: REPLACE {phone_names[k_existing]:<20} ({condition_names[condition]}) -> {phone_names[k_new]} (${C[k_new]:,})")
                    break
            else:
                # No phone at this location (removed)
                print(f"Location {j:2d}: REMOVE {phone_names[k_existing]:<20} ({condition_names[condition]})")
    
    # New installations
    print(f"\nNEW INSTALLATIONS:")
    new_installs = []
    for j in range(len(initial_phone_types), location_num):
        for k in range(4):  # Only new types can be installed
            if X[1, j, k].x > 0.5:
                new_installs.append((j, k))
                print(f"Location {j:2d}: Install {phone_names[k]:<20} (${C[k]:,})")
    
    # Summary
    print("\nSUMMARY:")
    print(f"  Existing locations kept as-is: {len(kept)}")
    print(f"  Existing locations replaced: {len(replaced)}")
    print(f"  New installations: {len(new_installs)}")
    print(f"  Total phones in period 1: {sum(X[1,j,k].x for j in range(location_num) for k in range(k_num)):.0f}")
    
    # Coverage statistics
    covered_count = sum(Y[i].x for i in range(demand_num))
    weighted_covered = sum(W[i] * Y[i].x for i in range(demand_num))
    
    print(f"\nCOVERAGE ACHIEVED:")
    print(f"  Demand points covered: {covered_count:.0f}/{demand_num} ({100*covered_count/demand_num:.1f}%)")
    print(f"  Weighted coverage: {weighted_covered:.1f}/{sum(W):.1f} ({100*weighted_covered/sum(W):.1f}%)")
    
    # Only sum costs for replacements (k != initial) or new locations
    inst_cost_existing = sum(
        X[1, j, k].x * C[k]
        for j in range(23)
        for k in range(4)
        if k != initial_phone_types[j]
    )
    inst_cost_new = sum(
        X[1, j, k].x * C[k]
        for j in range(23, location_num)
        for k in range(4)
    )
    inst_cost = inst_cost_existing + inst_cost_new
    
    penalty_cost = model.objVal - inst_cost
    print(f"\nCOST BREAKDOWN:")
    print(f"  Installation costs: ${inst_cost:,.2f}")
    print(f"  Penalty costs: ${penalty_cost:,.2f}")
    print(f"  Total: ${model.objVal:,.2f}")
```

### Visualisation

```python
# List to store row data
table_data = []

# Ensure the model is optimized
if model.status == GRB.OPTIMAL:
    
    # Iterate through all locations (j = 0 to 122)
    for j in range(location_num):
        
        # 1. Determine Initial State (t=0)
        # Existing locations (0-22) have data, New locations (23+) do not
        if j < len(initial_phone_types):
            phone_t0 = initial_phone_types[j]
            state_t0 = H[j]
        else:
            phone_t0 = "None"  # Or -1
            state_t0 = "None"  # Or -1
            
        # 2. Determine Optimal Decision (t=1)
        # Find which phone type k was selected (X[1,j,k] == 1)
        phone_t1 = None
        for k in range(k_num):
            if X[1, j, k].x > 0.5:  # Check if binary variable is 1
                phone_t1 = k
                break
        
        # 3. Calculate Specific Costs for this row
        cost_replacement = 0
        cost_penalty = 0
        
        if phone_t1 is not None:
            
            # Replacement Cost Calculation
            if j < len(initial_phone_types):
                # Existing: Only pay if we replace (k_new != k_old)
                if phone_t1 != phone_t0:
                    cost_replacement = C[phone_t1]
            else:
                # New Location: Always pay equipment cost
                cost_replacement = C[phone_t1]
                
            # Penalty Cost Calculation
            # Only applies if we keep an Old Phone (Type 4) at an existing location
            if j < len(initial_phone_types) and phone_t1 == 4:
                # Penalty = State * Lambda
                cost_penalty = state_t0 * lmbda
        
        # 4. Append row to data list
        table_data.append({
            "Location J": j,
            "Phone K (t = 0)": phone_t0,
            "State H (t = 0)": state_t0,
            "Phone K (t=1)": phone_t1,
            "Cost Replacement": cost_replacement,
            "Cost Penalty": cost_penalty
        })

    # Create and display DataFrame
    df_results = pd.DataFrame(table_data)
    print(df_results.head(23).to_string(index=False))
```

### Sensitivity Analysis

```python
# 1. Helper: automatically find the coverage constraint
def find_coverage_constraint(model, Y, W):
    """
    Try to locate the weighted coverage constraint of the form
        sum_i W[i] * Y[i] >= alpha * sum_i W[i]
    by searching for a constraint that:
      - has only Y[...] variables
      - has sense '>=' (encoded as '>')
      - has exactly len(W) variables in the row.
    """
    num_demands = len(W)
    y_names = {v.VarName for v in Y.values()}  # e.g. 'Y[0]', 'Y[1]', ...

    for c in model.getConstrs():
        row = model.getRow(c)
        # must involve exactly len(W) variables
        if row.size() != num_demands:
            continue
        # must be >= type constraint
        if c.Sense != '>':
            continue

        # check all vars are Y[...] variables
        all_y = True
        for k in range(row.size()):
            if row.getVar(k).VarName not in y_names:
                all_y = False
                break
        if not all_y:
            continue

        # this looks like our coverage constraint
        return c

    return None


# 2. Locate coverage constraint in your existing model
cov_constr = find_coverage_constraint(model, Y, W)
if cov_constr is None:
    raise RuntimeError("Could not automatically find the coverage constraint. "
                       "Check that Y is used only in the coverage constraint.")

print("Using coverage constraint:", cov_constr.ConstrName)

total_weight = sum(W)


# 3. Sensitivity on alpha for the FINAL model
alpha_values = [0.10, 0.15, 0.20, 0.25, 0.30, 0.35, 0.40, 0.45, 0.50, 0.55, 0.60, 0.65, 0.70, 0.75, 0.80]

alpha_results = []

for a in alpha_values:
    # 1) change RHS of coverage constraint to new alpha
    cov_constr.RHS = a * total_weight
    model.update()

    # 2) re-optimise
    model.optimize()

    if model.Status != GRB.OPTIMAL:
        alpha_results.append({
            "alpha": a,
            "total_cost": None,
            "coverage": None,
        })
        continue

    # 3) compute achieved coverage under this alpha
    covered_weight = sum(W[i] * Y[i].X for i in range(len(W)))
    coverage = covered_weight / total_weight if total_weight > 0 else 0.0

    alpha_results.append({
        "alpha": a,
        "total_cost": model.ObjVal,
        "coverage": coverage,
    })

df_alpha_final = pd.DataFrame(alpha_results)
display(df_alpha_final)


# 4. Plot sensitivity: cost + coverage vs alpha
fig, ax1 = plt.subplots(figsize=(8, 5))

# Left y-axis: total cost
line1, = ax1.plot(
    df_alpha_final["alpha"],
    df_alpha_final["total_cost"],
    marker="o",
    linewidth=2,
    label="Total Cost"
)
ax1.set_xlabel("Coverage requirement α")
ax1.set_ylabel("Total cost")

# Right y-axis: achieved coverage (%)
ax2 = ax1.twinx()
line2, = ax2.plot(
    df_alpha_final["alpha"],
    100 * df_alpha_final["coverage"],
    marker="s",
    linestyle="--",
    linewidth=2,
    label="Coverage Achieved (%)"
)
ax2.set_ylabel("Coverage achieved (%)")

# Combined legend
lines = [line1, line2]
labels = [l.get_label() for l in lines]
ax1.legend(lines, labels, loc="upper left")

plt.title("Sensitivity of Cost and Coverage to α (Final Model)")
plt.tight_layout()
plt.show()
```

# Extension 1: Installation and Infrastructure Costs + Maintenance Cost Instead of Penalty

```python
# Create model
model_ext3 = gb.Model('Phone Box - Complete Cost Model')
model_ext3.Params.OutputFlag = 1

# COSTS & PARAMETERS
INSTALL_TOWER_NEW, INSTALL_TOWER_REPLACE = 10000, 3000
INSTALL_WALL_NEW, INSTALL_WALL_REPLACE = 4000, 2000
ANNUAL_MAINTENANCE_COST = 1 * 52 * 30.62  # Hours * Weeks * Wage

# DECISION VARIABLES
X_ext3 = model_ext3.addVars(period_num, location_num, k_num, vtype=GRB.BINARY, name="X")
A_ext3 = model_ext3.addVars(demand_num, location_num, k_num, vtype=GRB.BINARY, name="A")
Y_ext3 = model_ext3.addVars(demand_num, vtype=GRB.BINARY, name="Y")

# Function to determine installation labor cost
def get_install_labor(k_new, k_old=None):
    if k_new in [0, 1]:  # Tower
        return INSTALL_TOWER_REPLACE if k_old in [0, 1] else INSTALL_TOWER_NEW
    elif k_new in [2, 3]:  # Wall
        return INSTALL_WALL_REPLACE if k_old in [2, 3, 4] else INSTALL_WALL_NEW
    return 0

# OBJECTIVE FUNCTION

# 1. Maintenance Cost (Only for keeping old phones at existing locations)
cost_maintenance = gb.quicksum(
    X_ext3[1, j, 4] * ANNUAL_MAINTENANCE_COST 
    for j in range(len(initial_phone_types))
)

# 2. Replacement Costs (Existing Locations: Equipment + Labor)
# Sum over existing locations (j < 23) and new phone types (k=0..3), if replacing (k != k_old)
cost_replacement = gb.quicksum(
    X_ext3[1, j, k] * (C[k] + get_install_labor(k, initial_phone_types[j]))
    for j in range(len(initial_phone_types))
    for k in range(4)
    if k != initial_phone_types[j]
)

# 3. New Installation Costs (Candidate Locations: Equipment + Labor)
# Sum over new locations (j >= 23) and new phone types (k=0..3)
cost_new_locs = gb.quicksum(
    X_ext3[1, j, k] * (C[k] + get_install_labor(k, None))
    for j in range(len(initial_phone_types), location_num)
    for k in range(4)
)

model_ext3.setObjective(cost_maintenance + cost_replacement + cost_new_locs, GRB.MINIMIZE)

# CONSTRAINTS

# 1. INITIAL STATE
model_ext3.addConstrs((X_ext3[0, j, initial_phone_types[j]] == 1 for j in range(len(initial_phone_types))), name="Init_Existing")
model_ext3.addConstrs((X_ext3[0, j, k] == 0 for j in range(len(initial_phone_types)) for k in range(k_num) if k != initial_phone_types[j]), name="Init_Existing_Zeros")
model_ext3.addConstrs((X_ext3[0, j, k] == 0 for j in range(len(initial_phone_types), location_num) for k in range(k_num)), name="Init_New_Zeros")

# 2. PERIOD 1: Max one phone per location
model_ext3.addConstrs((gb.quicksum(X_ext3[1, j, k] for k in range(k_num)) <= 1 for j in range(location_num)), name="One_Phone_Max")

# New locations cannot have old phones (k=4)
model_ext3.addConstrs((X_ext3[1, j, 4] == 0 for j in range(len(initial_phone_types), location_num)), name="No_Old_New_Locs")

# 3. COVERAGE
# Activation: A <= X
model_ext3.addConstrs((A_ext3[i, j, k] <= X_ext3[1, j, k] for i in range(demand_num) for j in range(location_num) for k in range(k_num)), name="Activation")

# Coverage: Sum(A) >= Y
model_ext3.addConstrs((gb.quicksum(A_ext3[i, j, k] for j in range(location_num) for k in range(k_num)) >= Y_ext3[i] for i in range(demand_num)), name="Coverage_Link")

# Radius Limits
model_ext3.addConstrs((A_ext3[i, j, k] == 0 for i in range(demand_num) for j in range(location_num) for k in range(k_num) if dist_matrix[i, j] > R[k]), name="Radius")

# Minimum Weighted Coverage
model_ext3.addConstr(gb.quicksum(W[i] * Y_ext3[i] for i in range(demand_num)) >= alpha * sum(W), name="Min_Coverage")

# 4. NO REMOVAL (Existing locations must keep a phone)
model_ext3.addConstrs((gb.quicksum(X_ext3[1, j, k] for k in range(k_num)) == 1 for j in range(len(initial_phone_types))), name="No_Removal")

# 5. NO DOWNGRADE
# For existing locations, X[1,j,k] = 0 if k > k_old (worse quality), unless k_old is 4 (Old Phone)
model_ext3.addConstrs(
    (X_ext3[1, j, k] == 0 
     for j in range(len(initial_phone_types)) 
     if initial_phone_types[j] < 4 
     for k in range(initial_phone_types[j] + 1, k_num)),
    name="No_Downgrade"
)

# Optimize
model_ext3.optimize()

# Print Results
if model_ext3.status == GRB.OPTIMAL:
    print(f"Total Cost: ${model_ext3.objVal:,.2f}")
```

```python
# Initialize accumulators
cost_equipment = 0
cost_installation = 0
cost_maintenance = 0

# Debug counters
count_kept_modern = 0
count_kept_old = 0
count_replaced = 0
count_new_build = 0

if model_ext3.status == GRB.OPTIMAL:
    for j in range(location_num):
        # Identify the active decision at location j
        active_k = None
        for k in range(5): # Explicitly check 0,1,2,3,4
            if X_ext3[1, j, k].x > 0.5:
                active_k = k
                break
        
        if active_k is not None:
            if j < len(initial_phone_types):
                k_existing = initial_phone_types[j]
                
                # Case A: KEPT SAME PHONE
                if active_k == k_existing:
                    if active_k == 4: # Keeping Old Phone
                        cost_maintenance += ANNUAL_MAINTENANCE_COST
                        count_kept_old += 1
                    else:             # Keeping Modern Phone (0-3)
                        # Cost is $0 - No Equipment, No Install
                        count_kept_modern += 1
                        
                # Case B: REPLACED PHONE
                else:
                    count_replaced += 1
                    # 1. Equipment Cost
                    cost_equipment += C[active_k]
                    
                    # 2. Installation Cost (Replacement Rates)
                    is_tower_new = (active_k in [0, 1])
                    was_tower_old = (k_existing in [0, 1])
                    
                    if is_tower_new:
                        if was_tower_old: cost_installation += INSTALL_TOWER_REPLACE
                        else:             cost_installation += INSTALL_TOWER_NEW
                    else: # New is Wall
                        # Old type 4 acts like wall mount for removal
                        was_wall_old = (k_existing in [2, 3, 4]) 
                        if was_wall_old:  cost_installation += INSTALL_WALL_REPLACE
                        else:             cost_installation += INSTALL_WALL_NEW

            else:
                count_new_build += 1
                # 1. Equipment Cost
                cost_equipment += C[active_k]
                
                # 2. Installation Cost (New Rates)
                if active_k in [0, 1]: # Tower
                    cost_installation += INSTALL_TOWER_NEW
                else:                  # Wall
                    cost_installation += INSTALL_WALL_NEW

    # Create Summary DataFrame
    total_project_cost = cost_equipment + cost_installation + cost_maintenance
    
    summary_data = {
        'Cost Category': [
            'Equipment (Infrastructure)', 
            'Installation Labor', 
            'Annual Maintenance', 
            'Total Project Cost'
        ],
        'Value ($)': [
            cost_equipment, 
            cost_installation, 
            cost_maintenance, 
            total_project_cost
        ]
    }

    df_summary_corrected = pd.DataFrame(summary_data)
    df_summary_corrected['Value ($)'] = df_summary_corrected['Value ($)'].apply(lambda x: f"${x:,.2f}")
    
    print("Decisions Breakdown (Debug):")
    print(f"  Kept Old (Maint):   {count_kept_old}")
    print(f"  Kept Modern ($0):   {count_kept_modern}")
    print(f"  Replaced:           {count_replaced}")
    print(f"  New Builds:         {count_new_build}")
    print(f"Calculated Total:     ${total_project_cost:,.2f}")
    print(f"Model Objective:      ${model_ext3.objVal:,.2f}")
    
    display(df_summary_corrected)
```
